```python
temp_path = "drive/MyDrive/mosiac/test_image.jpg"
root_folder = "/content/drive/MyDrive/mosiac/checkmark"
i = len(df)
for root,dirs,files in os.walk(root_folder):
  for f in files:
    img = Image.open(root_folder + "/" + f)
    new_size = (28,28)
    img = img.resize(new_size)
    img.save(temp_path)
    img = tf.image.decode_jpeg(tf.io.read_file(temp_path), channels=1)
    img = img.numpy()
    img = img.reshape(784,)
    img = list(255-img)
    #print(img)
    img = [1] + img
    #print(img)
    df.loc[i] = img
    i+= 1
    
```
